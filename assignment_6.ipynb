```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv("D:/Acadview training notes/read.csv")
plt.ylabel('Max tempreature')
plt.xlabel('Evaporation')
plt.plot(df["Evaporation"],df["MaxTemp"])
plt.show()

```

```python

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211)
plt.plot(t1,t1,'bo',label='t1')
plt.plot(t2,t2**2,'g',label='t2')
plt.title('subplots')
plt.legend()

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.grid(True)
plt.text(2.18,0.8,r'cos func')
plt.show()
```
